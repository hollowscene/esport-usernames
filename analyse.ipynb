### Setup

```python
import os

import pandas as pd
```

## Analysis

```python
csv_path = "data"
csv_file_name = "players_20241214163653.csv"
df = pd.read_csv(os.path.join(csv_path, csv_file_name))
```

```python
total_player_name_count = len(df["title"])
clean_player_name_count = len(df["cleantitle"].unique())
clean_ci_player_name_count = len(df["cleantitleupper"].unique())

print(f"Total player name count: {total_player_name_count:,}")
print(f"Unique player name count: {clean_player_name_count:,}")
print(f"Unique case-insensitive player name count: {clean_ci_player_name_count:,}")
```
