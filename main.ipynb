```python
from collections import Counter
import re

import main
```

```python
headers = {
    "User-Agent": "EsportUsernames/1.0 (https://github.com/hollowscene)"
}
```

```python
wiki = "leagueoflegends"
```

```python
%%time
players = main.get_player_pages(wiki=wiki, headers=headers)
player_names = [player["title"] for player in players]
```

i think i like the regex of at least 1 letter, space, left bracket, at least 1 letter, right bracket, ENDS. if it matches that, remove everything from the far right left bracket. JUST in case there's usernames with brackets in them

```python
# Testing
pattern = re.compile(r".+\ \(.+\)$") # Compile the regex pattern for efficiency

matches = main.find_matching_items(player_names, pattern)
print(matches)
print(len(matches))
```

```python
def isolate_username(player_name, regex_pattern):
    if re.search(regex_pattern, player_name):
        # strip everything from the last space + left bracket
        index = player_name.rfind(" (")
    
        # If a left bracket is found, slice the string up to and including this index
        if index != -1:
            return player_name[:index]
        else:
            raise ValueException("Regex pattern must be wrong")
            # return player_name
    else:
        return player_name
```

```python
cleaned_player_names = [isolate_username(player_name, pattern) for player_name in player_names]
```

```python
value_counts = Counter([player_name for player_name in cleaned_player_names])
```

```python
print(f"Total player name count in {wiki} wiki: {len(player_names):,}")
print(f"Unique player name count in {wiki} wiki: {len(value_counts):,}")
```
