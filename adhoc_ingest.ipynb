```python
url = f"https://liquipedia.net/{wiki}/api.php"

params = {
    "action": "query",
    "format": "json",
    "list": "categorymembers",
    "prop": "pageimages",
    "pageids": lucky_pageids,
}

response = requests.get(url, headers=headers, params=params)
data = response.json()
```
