Full wiki list copied from the **Available wikis** section of the [Liquipedia API documentation](https://api.liquipedia.net/documentation/api/v3).

```python
with open("wiki_list/wiki_list.txt", "r") as f:
    raw_wiki_list = f.read().split("\n")
```

Main wiki list created manually based on the [Liquipedia home page](https://liquipedia.net/).

*Last updated 15/07/2026. Home page order is "based on the total traffic of 1st half 2026 (January - June)".*

```python
main_wiki_list = [
    "dota2",
    "counterstrike",
    "valorant",
    "mobilelegends",
    "leagueoflegends",
    "rocketleague",
    "pubgmobile",
    "overwatch",
    "apexlegends",
    "fighters",
    "starcraft2",
    "pubg",
    "honorofkings",
    "ageofempires",
    "rainbowsix",
    "brawlstars",
    "marvelrivals",
    "fortnite",
    "callofduty",
    "smash",
    "warcraft",
    "starcraft",
    "worldoftanks",
    "easportsfc",
    "hearthstone",
    "wildrift",
    "heroes",
]
```

```python
main_wiki_dict = {}

for line in raw_wiki_list:
    display_name, wiki_name = line.split(" as ")
    if wiki_name in main_wiki_list:
        main_wiki_dict[wiki_name] = display_name

# Reorder dictionary based on Liquipedia home page order
main_wiki_dict = {key: main_wiki_dict[key] for key in main_wiki_list}
```

```python
with open("wiki_list/main_wiki_list.json", "w") as f:
    json.dump(main_wiki_dict, f, indent=4)
```
